```python
#Create a Basic ATM program that allows the users to check their Deposit money and Withdraw Money and ensure they don't withdraw more than they have.
# Initial balance
balance = 0.0

def check_balance():
    print(f"Your current balance is: ${balance:.2f}")

def deposit():
    global balance
    amount = float(input("Enter amount to deposit: $"))
    if amount > 0:
        balance += amount
        print(f"${amount:.2f} deposited successfully.")
    else:
        print("Invalid deposit amount. Please enter a positive value.")

def withdraw():
    global balance
    amount = float(input("Enter amount to withdraw: $"))
    if amount > balance:
        print("Insufficient funds. You cannot withdraw more than your current balance.")
    elif amount <= 0:
        print("Invalid withdrawal amount. Please enter a positive value.")
    else:
        balance -= amount
        print(f"${amount:.2f} withdrawn successfully.")

# ATM menu
while True:
    print("\n--- ATM Menu ---")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        check_balance()
    elif choice == "2":
        deposit()
    elif choice == "3":
        withdraw()
    elif choice == "4":
        print("Thank you for using the ATM. Goodbye!")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```

```python
#Write a Python Program to input the lenth of a series/list and create a list Dynalically, Hence search from the list using Linear Search.

n = int(input("Enter the number of elements in the list: "))

# Create an empty list
numbers = []

# Dynamically input the elements of the list
print("Enter the elements:")
for i in range(n):
    element = int(input(f"Element {i+1}: "))
    numbers.append(element)

# Input the element to search for
target = int(input("Enter the element to search for: "))

# Perform linear search
found = False
for i in range(n):
    if numbers[i] == target:
        print(f"Element {target} found at index {i}")
        found = True
        break

if not found:
    print(f"Element {target} not found in the list.")
```

```python
#Write a Python Program to create a Dictionary for every students data which may include the Name, Rollno, and Course ID. Hence create for the rolln and output the course ID

students = {}

# Input the number of students
n = int(input("Enter the number of students: "))

# Input data for each student
for i in range(n):
    print(f"\nEnter details for student {i+1}:")
    name = input("Enter name: ")
    roll_no = input("Enter roll number: ")
    course_id = input("Enter course ID: ")
    
    # Store the student's data in the dictionary using roll number as the key
    students[roll_no] = {
        "name": name,
        "course_id": course_id
    }

# Input the roll number to search for
search_roll_no = input("\nEnter the roll number to search for the course ID: ")

# Search for the roll number in the dictionary and output the course ID
if search_roll_no in students:
    course_id = students[search_roll_no]["course_id"]
    print(f"Course ID for roll number {search_roll_no} is: {course_id}")
else:
    print(f"No student found with roll number {search_roll_no}.")
```

```python
#Write a Python Program that displays a menu with item prices, and lets the user select the the item to add to their order and show the total bill after they finish.

menu = {
    "Burger": 5.99,
    "Pizza": 8.99,
    "Pasta": 7.49,
    "Salad": 4.49,
    "Soda": 1.99
}

# Function to display the menu
def display_menu():
    print("\nMenu:")
    for item, price in menu.items():
        print(f"{item}: ${price:.2f}")

# Function to take the user's order
def take_order():
    order = []
    total_bill = 0

    while True:
        display_menu()
        
        # Ask user to select an item or quit
        choice = input("\nEnter the item name to add to your order or type 'done' to finish: ").capitalize()
        
        if choice == 'Done':
            break
        elif choice in menu:
            order.append(choice)
            total_bill += menu[choice]
            print(f"{choice} added to your order. Current total: ${total_bill:.2f}")
        else:
            print("Invalid item, please choose from the menu.")
    
    return order, total_bill

# Main function to execute the order process
def main():
    print("Welcome to our restaurant!")
    
    order, total_bill = take_order()
    
    if order:
        print("\nYour order:")
        for item in order:
            print(f"- {item}")
        print(f"\nTotal Bill: ${total_bill:.2f}")
    else:
        print("No items were selected. Exiting...")

# Run the main function
main()
```

```python
#Write a Python Program to Create a function that outputs Factorials of a number without using Recursion.

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Input a number from the user
num = int(input("Enter a number: "))

# Call the factorial function and print the result
fact = factorial(num)
print(f"The factorial of {num} is {fact}.")
```

```python
#Write a Python Program to create a function that inputs any number of parameters and outputs Sum, Product and Average.

def calculate_operations(*args):
    # Calculate sum
    total_sum = sum(args)
    
    # Calculate product
    total_product = 1
    for num in args:
        total_product *= num
    
    # Calculate average
    average = total_sum / len(args) if args else 0
    
    return total_sum, total_product, average

# Input numbers from the user (can input multiple numbers separated by space)
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

# Call the function and get the results
sum_result, product_result, average_result = calculate_operations(*numbers)

# Output the results
print(f"Sum: {sum_result}")
print(f"Product: {product_result}")
print(f"Average: {average_result:.2f}")
```
